```python
#importing all the libraries 
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

# Print out a quick overview of the data
exists=os.path.isfile('./DSL-StrongPasswordData.csv')
if exists:
    data=pd.read_csv("DSL-StrongPasswordData.csv")
else :
    print("password_dataset not exist");  

dataset = data
dataset.drop(dataset.columns[[2]], axis=1, inplace=True)
grouped = dataset.groupby(['subject','sessionIndex']).sum()
df=grouped.T.groupby([s.split('.')[0] for s in grouped.T.index.values]).sum().T

#If to compare with new user
df['Speed'] = df['UD'] + df['H']
df.drop(['DD'], axis=1, inplace=True)
#Speed of each subject on basis of mean of H + UD key for each session
```

```python
#Speed of each subject on basis of mean of H + UD key for each subject
Speed= df.groupby(['subject']).mean()
Speed.head(10)
```

```python
#Calculating overall mean and standard deviation for each user
class UserCategory:    
    def __init__(self, subjects):
        self.user_scores = []
        self.mean_vector = []
        self.subjects = subjects
    
    def Operations(self):
        self.mean_vector = self.train.mean().values
        self.std_vector = self.train.std().values

    def evaluateMean(self):
        eerMean = [] 
        for subject in subjects:        
            genuine_user_data = data.loc[data.subject == subject,"H.period":"H.Return"]            
            self.train = genuine_user_data[:400]
            self.Operations()
            eerMean.append(self.mean_vector)
        return np.mean(eerMean)
    
    def evaluateStd(self):
        eerStd = [] 
        for subject in subjects:        
            genuine_user_data = data.loc[data.subject == subject,"H.period":"H.Return"]            
            self.train = genuine_user_data[:400]
            self.Operations()
            eerStd.append(self.std_vector)
        return np.std(eerStd)
    
subjects = dataset["subject"].unique()
print("Mean is",UserCategory(subjects).evaluateMean())  
print("Standard deviation",UserCategory(subjects).evaluateStd())
```
